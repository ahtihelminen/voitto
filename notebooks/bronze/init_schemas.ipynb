```python
%sql
-- 1. Create the Bronze Schema
CREATE SCHEMA IF NOT EXISTS voitto_bronze;

-- 2. Create the Raw Odds Table
-- Stores the full JSON response from TheOddsAPI
CREATE TABLE IF NOT EXISTS voitto_bronze.odds_api_raw (
    ingest_id STRING,          -- UUID for this specific fetch event
    ingest_timestamp TIMESTAMP,-- When we fetched it
    api_endpoint STRING,       -- e.g. 'v4/sports/basketball_nba/odds'
    raw_json STRING            -- The full unparsed JSON blob
)
USING DELTA
PARTITIONED BY (ingest_timestamp); 
-- Partitioning by timestamp helps clean up old raw data later if needed

-- 3. Create the Raw NBA Stats Table
-- Stores the full JSON response from the NBA API
CREATE TABLE IF NOT EXISTS voitto_bronze.nba_api_raw (
    ingest_id STRING,
    ingest_timestamp TIMESTAMP,
    game_date DATE,            -- The date of the games fetched
    data_type STRING,          -- e.g. 'scoreboard', 'boxscore', 'team_stats'
    raw_json STRING
)
USING DELTA
PARTITIONED BY (game_date);
```
